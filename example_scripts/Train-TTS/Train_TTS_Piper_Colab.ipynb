```python
from pydub import AudioSegment
from IPython.display import Audio

# Load the generated audio
sound = AudioSegment.from_wav("/content/test_output.wav")

# Slow down playback (0.8x speed = double the duration)
slowed_sound = sound._spawn(sound.raw_data, overrides={
    "frame_rate": int(sound.frame_rate * 0.5)
}).set_frame_rate(sound.frame_rate)

# Save slowed audio
slowed_path = "/content/test_output_slow.wav"
slowed_sound.export(slowed_path, format="wav")

# Play it in Colab
Audio(slowed_path)
```
